# Realizar la recopilacion de los años 2010 al 2022 de la base de datos de secuestros de la policia nacional
objetivos especificos:

**Objetivo de Negocio *
El objetivo del presente ensayo  es analizar y proponer estrategias efectivas de prevención de secuestros en Colombia, alineadas con el Objetivo de Desarrollo Sostenible número 16: Paz, Justicia e Instituciones Sólidas. El ensayo busca identificar enfoques innovadores y políticas públicas que puedan contribuir a reducir la incidencia de secuestros en un 30% para el año 2025, evaluando su impacto en la seguridad ciudadana y la construcción de una sociedad más justa y equitativa.

*Objetivo del Proyecto *
El objetivo específico del proyecto de investigación asociado  es llevar a cabo un análisis exhaustivo de los datos de secuestros presentados en Colombia desde el año 2010 hasta el 2022. A través de un enfoque interdisciplinario, se pretende identificar los patrones, factores y contextos que contribuyen al aumento o disminución de secuestros en diferentes regiones del país. Además, se busca proponer recomendaciones y estrategias basadas en evidencia para fortalecer las políticas de prevención, incluyendo la colaboración con organizaciones locales, la promoción de la educación comunitaria y la utilización de medios de comunicación como herramientas efectivas para sensibilizar a la sociedad.

*Objetivos Específicos del Proyecto de Investigación:*
1. Recopilar y analizar estadísticas detalladas de secuestros en Colombia durante el período 2010-2022.
2. Identificar patrones geográficos y temporales de secuestros para comprender su evolución a lo largo del tiempo y en diferentes áreas del país.
3. Examinar los factores socioeconómicos, políticos y culturales que pueden influir en la ocurrencia de secuestros.
4. Realizar un análisis comparativo con otras naciones latinoamericanas para contextualizar la situación de Colombia.
5. Entrevistar a expertos en seguridad, funcionarios gubernamentales y representantes de organizaciones civiles para obtener perspectivas múltiples sobre las políticas de prevención existentes.
6. Proponer estrategias específicas de prevención de secuestros basadas en las lecciones aprendidas de casos exitosos a nivel nacional e internacional.
7. Evaluar el impacto potencial de las estrategias propuestas en términos de reducción de secuestros y mejora de la seguridad ciudadana.

*Objetivos Específicos del proyecto *
1. Revisar la literatura académica y los informes pertinentes sobre la prevención de secuestros y la seguridad ciudadana en Colombia.
2. Analizar críticamente la eficacia de las políticas gubernamentales y las estrategias implementadas hasta la fecha.
3. Evaluar la relación entre los secuestros y factores como el conflicto armado, la desigualdad social y la falta de oportunidades económicas.
4. Identificar enfoques innovadores y buenas prácticas de prevención de secuestros en otras partes del mundo.
5. Proponer recomendaciones concretas para fortalecer las políticas de prevención de secuestros en Colombia, considerando la colaboración interinstitucional, la educación comunitaria y la movilización social.
6. Reflexionar sobre el papel de la sociedad civil, los medios de comunicación y las instituciones en la construcción de una cultura de paz y seguridad en el país.

```python
import pandas as pd

# Ruta al archivo Excel
archivo_excel = '/content/drive/MyDrive/Proyecto analitica /Secuestro_Andatos.xlsx'

# Definir nombres de columna manualmente
nombres_columna = ['ARMAS_MEDIOS', 'DEPARTAMENTO', 'MUNICIPIO','FECHA','DELITO','GENERO','*AGRUPA_EDAD_PERSONA', 'CODIGO_DANE', 'CANTIDAD']

# Cargar datos desde el archivo Excel con los nombres de columna definidos
df = pd.read_excel(archivo_excel, header=None, names=nombres_columna)

# Palabras clave que indican distintas edades
palabras_clave_edades = ['menor', 'adolescente', 'mayor']

# Función para verificar si el valor en la columna indica una cierta edad
def es_edad(valor):
    valor = valor.lower()
    for palabra in palabras_clave_edades:
        if palabra in valor:
            return True
    return False

# Contar filas con valores de distintas edades
cantidad_menores = df[df['*AGRUPA_EDAD_PERSONA'].apply(es_edad)].shape[0]

# Mostrar la cantidad de filas con valores de distintas edades
print(f"La cantidad de filas con valores de distintas edades es: {cantidad_menores}")
```
